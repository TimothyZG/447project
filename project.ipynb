```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
birds = pd.read_csv("bird.csv")
birds.head()
```

```python
# Extract the row
canada_goose_row = birds[birds["Species List"] == "Canada Goose"]

# Get the years and counts, safely converting to numeric
years = birds.columns[1:].astype(int)
counts_raw = canada_goose_row.iloc[0, 1:]

# Convert to numeric and drop NaNs
counts = pd.to_numeric(counts_raw, errors="coerce")

# Build tidy dataframe
df = pd.DataFrame({
    "Year": years,
    "Count": counts
})

# Drop any year with NaN count
df = df.dropna()

# Plot
sns.pointplot(data=df, x="Year", y="Count")
plt.xticks(rotation=45)
plt.title("Canada Goose Observations per Year")
plt.tight_layout()
plt.show()
```
